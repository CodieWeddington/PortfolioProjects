```python
# Codie Weddington
# Data Source: https://www.kaggle.com/datasets/danielgrijalvas/movies

# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# Read in the Data

df = pd.read_csv(r'C:\Users\Codie\Documents\Datasets\movies.csv')
```

```python
# Let's look at the data
df.head()
```

```python
#Let us check for any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# We will go ahead and remove any missing data

df = df.dropna()
```

```python
# Lets confirm the data is gone.

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# We will now confirm the data types

df.dtypes
```

```python
# We want to change the data type of some columns to make it easier to work with and make it look better

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.dtypes
```

```python
# Some of the years are off from the Released and the Year column, so we are going to create a correct year column

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
df
```

```python
# Let's add a little order to this data

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Let us check the correlation between some of the data

#Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title("Budget vs Gross Earnings")
plt.xlabel('Budget')
plt.ylabel('Gross Earnings')
plt.show
```

```python
# Plot Budget vs Gross using seaborn

sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws={"color": "red"}, line_kws={"color":"blue"})
```

```python
df.corr(method='pearson')
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Matrix for Numeric Features")
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show
```

```python
# Let's look at the top 15 companies by gross revenue

CompanyGrossSum = df.groupby('company')[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values('gross', ascending = False)[:15]

CompanyGrossSumSorted
```
